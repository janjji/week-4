```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("../Delivery truck trips data/delivery_trucks_cleaned.csv", low_memory=False)
```

```python
df
```

```python
df.dtypes
```

```python
# ── Datetime columns ──────────────────────────────────────────────────────────
for col in ['booking_date', 'actual_eta', 'trip_start', 'trip_end']:
    df[col] = pd.to_datetime(df[col], errors='coerce')
 
# ── Category columns (low-cardinality) ───────────────────────────────────────
for col in ['market_type', 'on_time', 'delay', 'vehicle_type']:
    df[col] = df[col].astype('category')
 
# ── String / ID columns ───────────────────────────────────────────────────────
for col in ['BookingID', 'vehicle_id', 'origin', 'driver_name',
            'customer_id', 'ping_time', 'planned_eta']:
    df[col] = df[col].astype('string')
 
# ── Float columns (already correct, just confirm) ─────────────────────────────
df['distance_km']    = df['distance_km'].astype('float64')
df['min_km_per_day'] = df['min_km_per_day'].astype('float64')
 
print(df.dtypes)
print(f"\nShape: {df.shape}")
```
