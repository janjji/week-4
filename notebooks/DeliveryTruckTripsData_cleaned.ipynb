```python
print("on_time values:", df['on_time'].unique())
print("delay values:", df['delay'].unique())
print("vehicle_type values:", df['vehicle_type'].unique())
```

```python
df['on_time'] = df['on_time'].map({'G': 'Y', 'unknown': 'N'})

print(df['on_time'].unique())
print(df['on_time'].value_counts())
```

```python
df['delay'] = df['delay'].map({'R': 'Y', 'unknown': 'N'})

print(df['delay'].unique())
print(df['delay'].value_counts())
```

```python
print("on_time Y:", (df['on_time'] == 'Y').sum())
print("on_time N:", (df['on_time'] == 'N').sum())
print("\ndelay Y:", (df['delay'] == 'Y').sum())
print("delay N:", (df['delay'] == 'N').sum())
```

```python
df['booking_date'] = pd.to_datetime(df['booking_date'], errors='coerce')

print(df['booking_date'].dtype)
print(df['booking_date'].head())
print(f"\nNull values: {df['booking_date'].isnull().sum()}")
```

```python
df['actual_eta'] = pd.to_datetime(df['actual_eta'], errors='coerce')

print(df['actual_eta'].dtype)
print(df['actual_eta'].head())
print(f"\nNull values: {df['actual_eta'].isnull().sum()}")
```

```python
df = df.dropna(subset=['actual_eta'])

print(f"Shape after dropping nulls: {df.shape}")
print(f"Null values remaining: {df['actual_eta'].isnull().sum()}")
```

```python
df['trip_start'] = pd.to_datetime(df['trip_start'], errors='coerce')

print(df['trip_start'].dtype)
print(df['trip_start'].head())
print(f"\nNull values: {df['trip_start'].isnull().sum()}")
```

```python
df.shape
```

```python
df['trip_end'] = pd.to_datetime(df['trip_end'], format='mixed', errors='coerce')

print(df['trip_end'].dtype)
print(df['trip_end'].head())
print(f"\nNull values: {df['trip_end'].isnull().sum()}")
```

```python
df = df.copy()

print(df['trip_end'].head(10))
print(df['trip_end'].unique()[:5])
```

```python
# Drop rows with null trip_end
df = df.dropna(subset=['trip_end'])

print(f"Shape after dropping nulls: {df.shape}")
print(f"Null values remaining: {df['trip_end'].isnull().sum()}")
```

```python
df
```

```python
df['origin'].value_counts()
```

```python
df['ping_time'] = df['ping_time'].astype(str).str.replace(r'\.0$', '', regex=True).str.strip()
df['planned_eta'] = df['planned_eta'].astype(str).str.replace(r'\.\d+$', '', regex=True).str.strip()

print(df['ping_time'].head())
print(df['planned_eta'].head(10))
```

```python
df['distance_km'] = df['distance_km'].round(2)
df['min_km_per_day'] = df['min_km_per_day'].round(2)

print(df['distance_km'].head())
print(df['min_km_per_day'].head())
```

```python
print(f"Unknown vehicle_type count: {(df['vehicle_type'] == 'unknown').sum()}")
print(f"Total rows: {df.shape[0]}")
```

```python
print(f"Unique drivers: {df['driver_name'].nunique()}")
print(f"Null driver_name: {df['driver_name'].isnull().sum()}")
print(f"\nUnique customers: {df['customer_id'].nunique()}")
print(f"Null customer_id: {df['customer_id'].isnull().sum()}")
```

```python
print(df.dtypes)
print(f"\nShape: {df.shape}")
print(f"\nNull values per column:")
print(df.isnull().sum())
```
