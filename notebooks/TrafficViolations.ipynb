```python
def categorize_violation(desc):
    desc = str(desc).upper()
    if 'SPEED' in desc or 'MPH' in desc:
        return 'SPEEDING'
    elif 'ALCOHOL' in desc or 'IMPAIRED' in desc or 'INFLUENCE' in desc or 'DUI' in desc:
        return 'DUI'
    elif 'SUSPENDED' in desc or 'REVOKED' in desc or 'LICENSE' in desc:
        return 'LICENSE_ISSUE'
    elif 'STOP' in desc or 'RED' in desc or 'SIGNAL' in desc or 'TRAFFIC CONTROL' in desc:
        return 'TRAFFIC_SIGNAL'
    elif 'REGISTRATION' in desc or 'PLATES' in desc or 'UNREGISTERED' in desc:
        return 'REGISTRATION'
    elif 'SEATBELT' in desc or 'SEAT BELT' in desc or 'RESTRAINED' in desc:
        return 'SEATBELT'
    elif 'PHONE' in desc or 'HANDHELD' in desc or 'TEXT' in desc:
        return 'DISTRACTED_DRIVING'
    elif 'RECKLESS' in desc or 'NEGLIGENT' in desc or 'CARELESS' in desc:
        return 'RECKLESS_DRIVING'
    elif 'LANE' in desc or 'TURN' in desc:
        return 'IMPROPER_LANE_USE'
    elif 'HEADLIGHT' in desc or 'LAMP' in desc or 'LIGHT' in desc:
        return 'EQUIPMENT_VIOLATION'
    else:
        return 'OTHER'

# Apply and replace Description
df['Description'] = df['Description'].apply(categorize_violation)

# Check the results
print(df['Description'].value_counts())
```
