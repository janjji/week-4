```python
dfexe = [
    'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16', 
    'Unnamed: 17', 'Unnamed: 18',
    'Area',
    'Difference',
    'Delivery Id' 
]
df = df.drop(columns=dfexe)
print(df.shape)
```

```python
df = df.dropna(subset=[
    'Origin_Location', 'Destination_Location', 'region',
    'created_at', 'On time Delivery', 'Customer_rating', 'condition_text'
])
print(f"Rows remaining: {len(df)}")
```

```python
df = df.copy()
df['Delivery_Time'] = df['Delivery_Time'].fillna(df['Delivery_Time'].median())
```

```python
df = df.dropna(subset=['actual_delivery_time'])
print(f"Rows remaining: {len(df)}")
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['created_at'] = pd.to_datetime(df['created_at'], errors='coerce')
df['actual_delivery_time'] = pd.to_datetime(df['actual_delivery_time'], errors='coerce')

print(df['created_at'].dtype)
print(df['actual_delivery_time'].dtype)
```

```python
df['hour'] = df['actual_delivery_time'].dt.hour
df['day_of_week'] = df['actual_delivery_time'].dt.dayofweek
df['month'] = df['actual_delivery_time'].dt.month
df['delivery_duration'] = (df['actual_delivery_time'] - df['created_at']).dt.total_seconds() / 3600
```

```python
df = df.drop(columns=['created_at', 'actual_delivery_time'])
```

```python
df['On time Delivery'] = df['On time Delivery'].astype(int)
df['Customer_rating'] = df['Customer_rating'].astype(int)
df['Fixed Costs'] = df['Fixed Costs'].astype(int)
df['Maintenance'] = df['Maintenance'].astype(int)
df['Delivery_Time'] = df['Delivery_Time'].astype(int)
df['delivery_duration'] = df['delivery_duration'].astype(int)
```

```python
df.dtypes
```

```python
print(df['condition_text'].value_counts())
print(df['region'].value_counts())
print(df['Origin_Location'].value_counts())
print(df['Destination_Location'].value_counts())
```

```python
def categorize_weather(condition):
    condition = str(condition).upper()
    if 'SUNNY' in condition or 'CLEAR' in condition:
        return 'CLEAR'
    elif 'CLOUDY' in condition or 'OVERCAST' in condition:
        return 'CLOUDY'
    elif 'RAIN' in condition or 'MIST' in condition or 'THUNDER' in condition:
        return 'RAINY'
    else:
        return 'OTHER'

df['condition_text'] = df['condition_text'].apply(categorize_weather)
print(df['condition_text'].value_counts())
```

```python
df['region'] = df['region'].str.strip().str.title()
df['region'] = df['region'].replace({'M': 'Maharashtra'})
print(df['region'].value_counts())
```

```python
def extract_state(location):
    location = str(location).upper()
    if 'TAMIL NADU' in location:
        return 'Tamil Nadu'
    elif 'KARNATAKA' in location:
        return 'Karnataka'
    elif 'PONDICHERRY' in location or 'PONDY' in location:
        return 'Pondicherry'
    elif 'MAHARASHTRA' in location or 'MUMBAI' in location:
        return 'Maharashtra'
    elif 'GUJARAT' in location:
        return 'Gujarat'
    elif 'HARYANA' in location:
        return 'Haryana'
    else:
        return 'OTHER'

df['Origin_Location'] = df['Origin_Location'].apply(extract_state)
df['Destination_Location'] = df['Destination_Location'].apply(extract_state)

print(df['Origin_Location'].value_counts())
print(df['Destination_Location'].value_counts())
```

```python
df = df.rename(columns={
    'Origin_Location': 'Origin',
    'Destination_Location': 'Destination',
    'region': 'Region',
    'On time Delivery': 'On_Time',
    'Customer_rating': 'Customer_Rating',
    'condition_text': 'Weather',
    'Delivery_Time': 'Delivery_Time',
    'hour': 'Hour',
    'day_of_week': 'Day_of_Week',
    'month': 'Month',
    'delivery_duration': 'Delivery_Duration'
})
```
